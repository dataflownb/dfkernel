The issue that we are seeing is that once making changes to the df variable creates a permanent change and to get the original data back, you must reload the data.

The following cells are an example of how we fixed this.

```python
import pandas as pd

df = pd.read_csv("../examples/fifa17-top20-women-fixed.txt",sep="-",header=None)
```

```python
df = df$67068118.rename(columns={0: "Name", 1: "Country", 2: "Rating"})
```

```python
df = df$87b0107b.assign(Name=df$87b0107b["Name"].str.strip(), 
                 Country=df$87b0107b["Country"].str.strip())
```

We showed above that you can still access the original dataframe after editing and creating "new" ones. Because of the nature of Dataflow Notebooks being output based, we are able to reference "older versions" of variables.
